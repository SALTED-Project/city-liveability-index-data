```python
existing_entity_id_list = {}
def save_main_indices(df, entity_type, entity_ID, location, date_format, sub_index_prefix):
    counter= 0
    example_flag=True

    
    date_time_list= df.date_time.unique()
    
    for date_time in date_time_list:
        
        entity_obj = Entity(entity_type, entity_ID, ["https://raw.githubusercontent.com/smart-data-models/data-models/master/context.jsonld"]) # (re-)initialize the entity

        
        date_time_data = df[df['date_time'] == date_time]
        if(len(date_time_data) ==0):
            continue
        related_hexagon_id_list = date_time_data.hexagon_id.unique()
        if(len(related_hexagon_id_list)):
            print("Warning: There is no related hexagon for the general index. This should not happen")
        index_value = date_time_data['current_annual_mean'].mean()
        
        # add properties
        # geoproperty
        entity_obj.add_geoprop("location", location, observedAt= datetime.strptime(str(date_time_data.iloc[0].date_time), date_format))
        # value
        entity_obj.add_prop('current_annual_mean', index_value, observedAt= datetime.strptime(str(date_time_data.iloc[0].date_time), date_format))
        entity_obj.add_prop('data', index_value, observedAt= datetime.strptime(str(date_time_data.iloc[0].date_time), date_format))
        # relations --> related hexagons
        for hexagon_id in related_hexagon_id_list:
            related_id =  sub_index_prefix  + str(hexagon_id)
            entity_obj.add_rel('providedBy', related_id, observedAt= datetime.strptime(str(date_time_data.iloc[0].date_time), date_format))
            publish(entity_obj.to_dict()) # publish the entity
        if example_flag:
            print('INFO: Example data from the dataset to be published - as an example on how data looks.')
            print(entity_obj)
            example_flag=False

        print('Info: Dataset is published successfully with entity ID: ' + entity_ID) 
```
